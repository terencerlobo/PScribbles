```python
import math
import numpy as np
```

```python
t = np.arange(0,2.5,0.1)
y1 = np.sin(math.pi*t)
y2 = np.sin(math.pi*t + math.pi/2)
y3 = np.sin(math.pi*t - math.pi/2)
plt.plot(t,y1,'b*')
plt.plot(t,y2,'g^')
plt.plot(t,y3,'rs')
```

```python
t = np.arange(0,5,0.1)
y1 = np.sin(2*np.pi*t)
y2 = np.cos(2*np.pi*t)

plt.subplot(2,1,1)
plt.plot(t,y1,'b-')
plt.subplot(2,1,2)
plt.plot(t,y2,'r--')
```

```python
plt.subplot(1,2,1)
plt.plot(t,y1,'b-')
plt.subplot(1,2,2)
plt.plot(t,y2,'r--')
```
