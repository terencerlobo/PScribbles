```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = {'series1': [1,3,4,3,5],
        'series2': [2,4,5,2,4],
        'series3': [3,2,3,1,3] }

df = pd.DataFrame(data)


x = np.arange(5)
plt.axis([0,5,0,7])
plt.plot(x,df.series1)
plt.plot(x,df.series2)
plt.plot(x,df.series3)
plt.legend(data, loc=2)
print(plt, " Here is the plt.")
```
