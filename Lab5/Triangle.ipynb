```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = {'red-dotted': [20,40,10],
        'green-dashed': [40,10,30]}
df = pd.DataFrame(data)
x = [10,20,30]


```

```python

plt.axis([10,30,10,40])
plt.title("My First Plot")
plt.xlabel("This is X axis")
plt.ylabel("This is y axis")
plt.plot(x,df.get("red-dotted"), color='r', linestyle = 'dotted', linewidth=5)
plt.plot(x,df.get("green-dashed"), color='g', linestyle = '--', linewidth=5)
plt.legend(data, loc=2)
```
